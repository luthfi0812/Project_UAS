```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r'C:\Users\Lenovo\DASHPRO24H\player_valuations.csv')

df['date'] = pd.to_datetime(df['date'], errors='coerce')
df = df.dropna(subset=['date'])
df['year'] = df['date'].dt.year

league_counts = df['player_club_domestic_competition_id'].value_counts().head(10)

plt.figure(figsize=(10, 6))
plt.barh(league_counts.index, league_counts.values, color='green')
plt.title('Top 10 Liga Berdasarkan Jumlah Pemain')
plt.xlabel('Jumlah Pemain')
plt.ylabel('Liga')
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()

avg_market_value = df.groupby('year')['market_value_in_eur'].mean().reset_index()

plt.figure(figsize=(10, 6))
plt.plot(avg_market_value['year'], avg_market_value['market_value_in_eur'], marker='o', color='steelblue')
plt.title('Rata-rata Nilai Pasar Pemain per Tahun')
plt.xlabel('Tahun')
plt.ylabel('Rata-rata Nilai Pasar (EUR)')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```
